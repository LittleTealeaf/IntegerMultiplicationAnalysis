```python
import pandas as pd

data = pd.read_csv('results.csv')
data[['Default','Karatsuba','Java']] = data[['Default','Karatsuba','Java']] * 1e-9
data
```

```python
data.plot(x='Digits',y=['Default','Karatsuba','Java'],title="Time Spent vs Digit Count",ylabel="Time (seconds)")
data.plot(x='Digits',y=['Default','Karatsuba','Java'],title="Log Time Spent vs Log Digit Count",ylabel="Log Time (log(seconds))",xlabel="Log Digits (log(seconds))",loglog=True)
```
